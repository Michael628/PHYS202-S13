```python
import numpy as np
import matplotlib.pyplot as plt

from mpl_toolkits.mplot3d import Axes3D

class Event(object):

    def __init__(self, id):
        self.id = id
        self.edat = []
        self.traj = []
        self.orphans = []
                                       
    def getID(self):
        return self.id
    
    def getData(self):
        return self.edat
                    
    def getTrajectories(self):
        return self.traj
    
    def getOrphans(self):
        return self.orphans
            
    def printData(self):
        for v in self.edat:
            print v.toString()
    
    def addVoxel(self, d):
        self.edat.append(Voxel([int(d[0]), int(d[1]), int(d[2]), int(d[3])] , int(d[4])))
            
    def popNeighbors(self, vox):
        poplist = []
        for v in listNeighbors(self.edat, vox):
            poplist.append(self.edat.pop(self.edat.index(v)))
        return poplist
    
    def makeTrajectories(self, gradthresh, dirthresh):
        self.edat = sorted(self.edat, key = lambda v: v.getVal(), reverse=True)
        
        maxlen = len(self.edat)
        x = 0
        
        while len(self.edat) > 0:
            vox = self.edat.pop(0)
            newTraj = Trajectory(vox)
            self.traj.append(newTraj)
            
            #breaks when there are no more candidates being generated
            while True:
                
                candidates = []
                rejects = []
                chosen = []
                
                neigh = self.popNeighbors(vox)
                candidates.extend(neigh)
                
                grad = getGradient(vox, neigh)
                
                for v in candidates:
                    #print v.toString()
                    #print grad[v]
                    thresh = vox.getVal() * gradthresh
                    if grad[v][1] <= thresh:
                        if newTraj.checkDir(grad[v][0], dirthresh):
                            #voxel matches trajectory
                            chosen.append(v)
                        else:
                            rejects.append(v)
                    else:
                        #print "reject grd", grad[v][1], thresh
                        rejects.append(v)
                
                #no new chosen, time to stop
                if len(chosen) == 0:
                    self.edat.extend(rejects)
                    break
                
                #find the lowest gradient of the neighbors. that will be our new tail
                min = 1000000
                lowestgrad = None
                for v in chosen:
                    if grad[v][1] < min:
                        min = grad[v][1]
                        lowestgrad = v        
                vox = chosen.pop(chosen.index(lowestgrad))
                newTraj.addSpine(vox)
                
                #add the rest of the voxels that match gradient and direction but are not best
                for v in chosen:
                    newTraj.addFlesh(v)
                
                #recycle unused voxels            
                self.edat.extend(rejects)
            #end of chained tail loop
            
            self.edat = sorted(self.edat, key = lambda v: v.getVal(), reverse=True)
            
        #get the orphans
        for t in reversed(self.traj):
            if len(t.getSpine()) == 1:
                self.orphans.extend(t.getMembers())
                self.traj.remove(t)
        
    def mergeTrajectories(self, mergethresh):
        counter = 0
        numtraj = len(self.traj)
        for x in range(numtraj):
            t2 = self.traj.pop(0)
            tdir = t2.getDir()
            
            merged = False
            for i in range(numtraj):
                if i >= len(self.traj):
                    break
                t = self.traj[i]
                if t.checkDir(tdir, .05) and (len(listNeighbors(t.getMembers(), t2.getTail())) > 0 or len(listNeighbors(t.getMembers(), t2.getHead()))):
                    t.merge(t2)
                    merged = True
                    counter += 1
                    break
            if not merged:
                self.traj.append(t2)

        numtraj = len(self.traj)
        for x in range(numtraj):
            t2 = self.traj.pop(0)
            tdir = t2.getDir()
            
            merged = False
            for i in range(numtraj):
                if i >= len(self.traj):
                    break
                t = self.traj[i]
                if t.checkDir(tdir, mergethresh) and (len(listNeighbors(t.getMembers(), t2.getTail())) > 0 or len(listNeighbors(t.getMembers(), t2.getHead()))):
                    t.merge(t2)
                    merged = True
                    counter += 1
                    print "merged:", i,x
                else:
                    print "failed to merge:", i,x
            if not merged:
                print "failed to merge:", i,x
                self.traj.append(t2)
        return counter
        
class Trajectory(object):
    def __init__(self, vox):
        self.du = 0.
        self.dv = 0.
        self.dw = 0.
        self.directions = []
        self.flesh = []
        self.spine = [vox]
        self.tail = vox
    
    def addFlesh(self, vox):
        self.flesh.append(vox)
        
    def addSpine(self, vox):
        self.spine.append(vox)
        self.tail = vox
        self.directions.append(getGradient(self.spine[-2], [self.spine[-1]])[self.spine[-1]][0])
        self.du = sum([v[0] for v in self.directions]) / float(len(self.directions))
        self.dv = sum([v[1] for v in self.directions]) / float(len(self.directions))
        self.dw = sum([v[2] for v in self.directions]) / float(len(self.directions))
    
    def merge(self, tra):
        for v in tra.getSpine():
            self.addSpine(v)
        for v in tra.getFlesh():
            self.addFlesh(v)

    def getDir(self):
        return self.du,self.dv,self.dw
        
    def getTail(self):
        return self.spine[-1]
    
    def getHead(self):
        return self.spine[0]
    
    def getFlesh(self):
        return self.flesh
    
    def getSpine(self):
        return self.spine
    
    def getMembers(self):
        all = []
        all.extend(self.flesh)
        all.extend(self.spine)
        return all
    
    def checkDir(self, (vdu, vdv, vdw), dirthresh):
        
        if self.du == 0 and self.dv == 0 and self.dw ==0:
            return True
        
        err = (self.du - vdu)**2 + (self.dv - vdv)**2 + (self.dw - vdw)**2
        if err > dirthresh:
            #print "reject dir:", err
            return False
        return True
    
    def toString(self):
        s = str(self.du) + ", " + str(self.dv) + ", " + str(self.dw) + "\n"
        s += str(len(self.spine)) + " , " + str(len(self.flesh)) + "\n---"
        
        """
        for v in self.spine:
            s += v.toString() + ",  "
        s += "\n---\n"
        for v in self.flesh:
            s += v.toString() + ",  "
        s += "\n.............\n"
        """
        return s
        
class Voxel(object):
    def __init__(self, idarray, adc):
        self.id = (idarray[0], idarray[1], idarray[2], idarray[3])
        self.adc = adc
        
    def getID(self):
        return self.id
    
    def getVal(self):
        return self.adc
    
    def toString(self):
        return str(self.id) + "  " + str(self.adc)
          
def listNeighbors(li, vox):
        ne = []
        q = vox.getID()
        nmap = {
            (q[0], q[1], q[2]-1, q[3]), (q[0], q[1]-1, q[2], q[3]), (q[0], q[1]-1, q[2]+1, q[3]),
            (q[0], q[1], q[2]+1, q[3]), (q[0], q[1]+1, q[2]+1, q[3]), (q[0], q[1]+1, q[2], q[3]),
            
            (q[0], q[1], q[2], q[3]-1), (q[0], q[1], q[2]-1, q[3]-1), (q[0], q[1]-1, q[2], q[3]-1),
            (q[0], q[1]-1, q[2]+1, q[3]-1), (q[0], q[1], q[2]+1, q[3]-1), (q[0], q[1]+1, q[2]+1, q[3]-1),
            (q[0], q[1]+1, q[2], q[3]-1),
            
            (q[0], q[1], q[2], q[3]+1), (q[0], q[1], q[2]-1, q[3]+1), (q[0], q[1]-1, q[2], q[3]+1),
            (q[0], q[1]-1, q[2]+1, q[3]+1), (q[0], q[1], q[2]+1, q[3]+1), (q[0], q[1]+1, q[2]+1, q[3]+1),
            (q[0], q[1]+1, q[2], q[3]+1)
            }
        for v in li:
            if v.getID() in nmap:
                unique = True
                for n in ne:
                    if v.getID() == n.getID():
                        unique = False
                        print "**Error: Duplicate voxel**", v.getID()
                if unique:        
                    ne.append(v)    
        return ne

def voxelsToArray(li):
    le = len(li)
    earr = [np.zeros(le), np.zeros(le), np.zeros(le), np.zeros(le), np.zeros(le)]
    for i, v in enumerate(li):
        earr[0][i] = v.getID()[0]
        earr[1][i] = v.getID()[1]
        earr[2][i] = v.getID()[2]
        earr[3][i] = v.getID()[3]
        earr[4][i] = v.getVal()
    return earr
    
def plotEvent(ev, filename):

    fig = plt.figure(1)
    ax = fig.add_subplot(1,1,1, projection = '3d')
    
    colors = ["red", "orange", "yellow", "green", "blue", "purple", "pink"]
    for i, t in enumerate(ev.getTrajectories()):
        evda = voxelsToArray(t.getSpine())
        
        #tcolor = int(i%2), int(i%3%2), int(i%5%2), 1
        #if tcolor == (1,1,1,1):
        #   tcolor = .5,.5,.5,1
        if i > 6:
            tcolor = 'gray'
        else :
            tcolor = colors[i]
        
        p = ax.scatter(evda[1], evda[2], evda[3], s=1.5*evda[4], linewidth=0, color=tcolor)
    
    #plt.xlim(20, 40)
    #plt.ylim(20, 40)
    
    
    ax.set_xlabel('Row')
    ax.set_ylabel("Column")
    ax.set_zlabel('Bucket')
    
    #plt.savefig(filename)
    plt.show()
    plt.close()
    
    
def getGradient(vox, neigh):
    grad = dict()
    for v in neigh:
        
        dE = vox.getVal() - v.getVal()
        
        du = vox.getID()[1] - v.getID()[1]
        dv = vox.getID()[2] - v.getID()[2]
        dw = vox.getID()[3] - v.getID()[3]
        
        grad[v] = (du,dv,dw), dE
    #print "------"
    #print len(grad)
    #print "---..--"
    return grad
        
#setup and parse
evlist = []
evcounter = 0
datain = open('niffte_data.txt', 'r')

for i, line in enumerate (datain):
    if line.startswith('#'):
        evlist.append(Event(evcounter))
        evcounter += 1
    else:
        s = np.fromstring(line, sep=" ")
        evlist[-1].addVoxel(s)
        
datain.close()

```
