Publication ready plot template
---
Simple graphing template for matplotlib functions and data plots.

### Line plots

```python
#some sample data for plots
y = arange(-5., 5.+.01, .01)
E = cosh(y)
mp = sinh(y)
```

```python
#object oriented setup
fig = plt.figure()
ax = fig.add_subplot(1,1,1)

#main plot parameters go here
plt.plot(y, E, linestyle = '-', color = 'red', linewidth = 1., label = "$foo$")
plt.plot(y, mp, label = "$bar$")

#title
ax.set_title('stub title')

#spines, aka axis
ax.spines['left'].set_position('center')
ax.spines['right'].set_color('none')
ax.spines['bottom'].set_position('center')
ax.spines['top'].set_color('none')

#tick marks
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

#axis labels
plt.xlabel("$stub$")
plt.ylabel("$stub()$")

#legend location
plt.legend(loc='best', fontsize = 9)

#show the plot. ties focus outside of ipython.
plt.show()
```

### Subplots

```python
fig = plt.figure()
#subplot command: 1 row, 2 columns, 1 st plot
ax = plt.subplot(1, 2, 1)

#main plot parameters go here
plt.plot(y, E, linestyle = '-', color = 'red', linewidth = 1., label = "$foo$")

#title
ax.set_title('stub title')

#spines, aka axis
ax.spines['left'].set_position('center')
ax.spines['right'].set_color('none')
ax.spines['bottom'].set_position('center')
ax.spines['top'].set_color('none')

#tick marks
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

#axis labels
plt.xlabel("$stub$")
plt.ylabel("$stub()$")

#subplot command: 1 row, 2 columns, 2 nd plot
ax = plt.subplot(1, 2, 2)
plt.plot(y, mp, label = "$bar$")

#title
ax.set_title('stub title')

#spines, aka axis
ax.spines['left'].set_position('center')
ax.spines['right'].set_color('none')
ax.spines['bottom'].set_position('center')
ax.spines['top'].set_color('none')

#tick marks
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

#axis labels
plt.xlabel("$stub$")
plt.ylabel("$stub()$")


plt.show()
```
