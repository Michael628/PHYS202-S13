#### 2. Particle in a Box

$P(a,b) = \displaystyle \int _a^b|\psi(x)|^2dx $

$\psi(x) = \sqrt{ \frac{2}{L} } \sin(\frac{N \pi x}{L}) $

```python
import scipy.integrate as integrate

L = 1
ra = np.arange(0,L + .01,.01)

f = lambda x, n: ((2/L)**.5 * np.sin((n*np.pi*x)/L))**2

psisq, err = integrate.quad(f,1./3.*L, .5*L, args=(1))
print "N = 1, probability is", psisq, " with error =", err
psisq, err = integrate.quad(f,1./3.*L, .5*L, args=(2))
print "N = 2, probability is", psisq, " with error =", err
```

```python
#using trapz instead
#n=1
dx = .0000005
ra = np.arange(1./3.*L,.5*L,dx)
ftrap = np.zeros(ra.size)
for i,j in enumerate(ra):
    ftrap[i] = f(j,1)
    
psisq = integrate.trapz(ftrap, dx=dx)
err = abs(psisq - .304498890522)
    
print "N = 1, probability is", psisq, " with error =", err
print 1/dx, "bins"

#n=2 
dx = .001
ra = np.arange(1./3.*L,.5*L,dx)
ftrap = np.zeros(ra.size)
for i,j in enumerate(ra):
    ftrap[i] = f(j,2)
    
psisq = integrate.trapz(ftrap, dx=dx)
err = abs(psisq - .0977505547389)
    
print "N = 2, probability is", psisq, " with error =", err
print 1/dx, "bins"
```

(a) N = 1, probability is 0.304498890522; N = 2, probability is 0.0977505547389

Quad quickly and easily gives a very accurate answer. Trapz is slow and inaccurate. It required 2 million bins to reach the desired accuracy for N=1, though only 1000 for N=2.

```python
def H(n,x):
    """
    Recursively generates (physicist's) hermite polynomials 
    in: int n, the polynomial number; array x, a function-like to act on
    out: array, function-like
    """
    if n < 0:
        print "error invalid n"
        return 1
    elif n == 0:
        return 1
    elif n == 1:
        return 2*x
    elif n > 1:
        #recursion is so elegant even if it isn't the fastest way
        return 2*x*H(n-1,x) - 2*(n-1)*H(n-2,x)
'''
    Testing Hermite polynomials
    print H(2,1), H(2,2), H(1,2), H(3,1),H(4,1), H(5,3)
'''
print
```

#### 4. Monte Carlo Integration

```python
#hit or miss method
def hmMC(f, N, xmin, xmax):
    """
    Monte Carlo integration method. Integrates by comparing random values 
    to function's values across a region, then counting the number of hits
    versus the number of misses to compute the integral
    in: function f, number of random points to sample N, limits xmin,xmax
    out: float integral value
    """
    x = arange(xmin, xmax+.001, .001)

    #find maximum for bb
    subx = x[logical_and((x > xmin), (x < xmax))]
    ymax = max(f(subx))*1.05
    ymin = 0.

    samples_x = xmin + (xmax - xmin) * random_sample(N)
    samples_y = ymin + (ymax - ymin) * random_sample(N)
    
    newmask = (samples_y < f(samples_x))
    return (xmax - xmin) * (ymax - ymin) * sum(newmask)/newmask.size

f = lambda x: 4*(1-x**2)**.5

print "Hit or Miss MC Integral =", hmMC(f, 10000, 0, 1)
```

(b) Calculate the difference between your result from part (a) and the exact result π for the N values from part (a).

```python
Imc = hmMC(f, 10000, 0, 1)
Iexact = np.pi
Ierr = 100*(Iexact - Imc) / Iexact

print "Integral_exact =", Iexact
print "Integral_MC =", Imc
print "Error of MC =", Ierr
```

```python
def MCerror(n):
    """
    private function, wrapper for hmMC integration
    """
    f = lambda x: 4*(1-x**2)**.5
    xmin = 0
    xmax = 1
    Imc = hmMC(f, n, xmin, xmax)
    Iexact = np.pi
    return np.abs((Iexact - Imc) / Iexact)

maxN = 1E4+10
winnow = 1
n = np.arange(1,int(maxN), 1)

nvals = np.arange(1, maxN, 1)
error = np.zeros(maxN-1)

for i,n in enumerate(n):
    error[i] = MCerror(n)
#print error[1], error[10], error[100], error[1000], error[1E4]
```

The upper bound of the error for large N is very roughly $\frac{1}{2 \ln(N)}.$

The hit or miss MC algorithm achieves one digit of accuracy at about N=100 and doesn't achieve two digits of accuracy even at N=10000. (Using approximate upper bound)

(d) Using Sample Mean MC

```python
#sample mean method
def smMC(f, N, xmin, xmax):
    """
    Improved Monte Carlo integration method. Integrates by slicing a region in
    one dimension and finding the mean of the integral in that dimension
    by sampling random points and comparing them to the function's value.
    The total integral is the average of those means.
    in: function f, number of random points to sample N, limits xmin,xmax
    out: float integral value
    """
    x = arange(xmin, xmax+.001, .001)

    #find maximum for bb
    subx = x[logical_and((x > xmin), (x < xmax))]
    ymax = max(f(subx))*1.05
    ymin = 0.

    samples_x = xmin + (xmax-xmin)*random_sample(N)
    guess_y = f(samples_x)
    avg = np.average(guess_y)
    return avg*(xmax-xmin)

f = lambda x: 4*(1-x**2)**.5

Imc = smMC(f, 10000, 0, 1)
Iexact = np.pi
Ierr = abs(100*(Iexact - Imc) / Iexact)

print "Integral_exact =", Iexact
print "Integral_MC =", Imc
print "Error of MC =", Ierr
```

```python
def smMCerror(n):
    """
    private function, wrapper for smMC integration
    """
    f = lambda x: 4*(1-x**2)**.5
    xmin = 0
    xmax = 1
    Imc = smMC(f, n, xmin, xmax)
    Iexact = np.pi
    return np.abs((Iexact - Imc) / Iexact)

maxN = 1E4+10
winnow = 1
n = np.arange(1,int(maxN), 1)

nvals = np.arange(1, maxN, 1)
error = np.zeros(maxN-1)

for i,n in enumerate(n):
    error[i] = smMCerror(n)
#print error[1], error[10], error[100], error[1000], error[1E4]
```

```python
fig = plt.figure()
ax = plt.subplot(111)

plt.plot(nvals, error, linewidth=0, marker='.', color=(.6,0,0,.5))
ax.set_yscale('log')
ax.set_xscale('log')

plt.xlabel("N (log scale)")
plt.ylabel("Error (log scale)")
plt.title("Error of Sample Mean MC")
plt.axis('tight')
plt.show() 
```

The Sample Mean method is clearly more accurate. One decimal of accuracy is achieved around N=10, two digits are achieved at about N = 10000. So the error's functional dependence on N is roughly $\frac{1}{\sqrt N} $. (Using approximate upper bound)

```python
import scipy.integrate as integrate

x0 = -.1; x1 = .1
y0 = -.1; y1 = .1
z0 = -.5; z1 = .5

y0f = lambda y: y0; y1f = lambda y: y1
x0f = lambda x,y: x0; x1f = lambda x,y: x1

#mass
rho = lambda x,y,z: 100*(y+.1 + z**2)

print integrate.tplquad(rho, z0, z1, y0f, y1f, x0f, x1f)
```

```python
#center of mass
def fx(x,y,z):
    return x * 100*(y+.1 + z**2)
def fy(x,y,z):
    return y * 100*(y+.1 + z**2)
def fz(x,y,z):
    return z * 100*(y+.1 + z**2)

coords = (integrate.tplquad(fx, z0, z1, y0f, y1f, x0f, x1f)[0], integrate.tplquad(fy, z0, z1, y0f, y1f, x0f, x1f)[0], integrate.tplquad(fz, z0, z1, y0f, y1f, x0f, x1f)[0])
print coords
print "error:", integrate.tplquad(fy, z0, z1, y0f, y1f, x0f, x1f)[1]
```

```python
#moment of inertia
def fx(x,y,z):
    return ((y**2+z**2)**.5)**2 * 100*(y+.1 + z**2)
def fy(x,y,z):
    return ((x**2+z**2)**.5-0.013333333)**2 * 100*(y+.1 + z**2)
def fz(x,y,z):
    return ((x**2+y**2)**.5)**2 * 100*(y+.1 + z**2)

coords = (integrate.tplquad(fx, z0, z1, y0f, y1f, x0f, x1f)[0], integrate.tplquad(fy, z0, z1, y0f, y1f, x0f, x1f)[0], integrate.tplquad(fz, z0, z1, y0f, y1f, x0f, x1f)[0])
print coords
print "error:", integrate.tplquad(fy, z0, z1, y0f, y1f, x0f, x1f)[1]
```
