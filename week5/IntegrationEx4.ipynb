$ E(x) = \displaystyle\int_0^xe^{-t^2}dt $

(a)

```python
import scipy.integrate as integrate

ra = np.arange(0,3.1,.1)
f = lambda t: np.exp(-t**2)
Ex = np.zeros(3.1/.1)
for i, v in enumerate(ra):
    Ex[i] = integrate.quad(f,0,v)[0]
print Ex
```

```python
plot(ra, Ex)
plt.ylabel(r"$E(x)$")
plt.xlabel("$x$")
show()
```

2\. $P(a,b) = \displaystyle \int _a^b|\psi(x)|^2dx $

$\psi(x) = \sqrt{ \frac{2}{L} } \sin(\frac{N \pi x}{L}) $

```python
import scipy.integrate as integrate


L = 1
ra = np.arange(0,L + .01,.01)

f = lambda x, n: ((2/L)**.5 * np.sin((n*np.pi*x)/L))**2

psisq, err = integrate.quad(f,1./3.*L, .5*L, args=(1))
print "N = 1, probability is", psisq, " with error =", err
psisq, err = integrate.quad(f,1./3.*L, .5*L, args=(2))
print "N = 2, probability is", psisq, " with error =", err
```
