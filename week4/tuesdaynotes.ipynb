Electric Diapole Potential
====

$ V(x,y) = \displaystyle\frac{kq}{\sqrt{x^2 + (y-d)^2}} - \frac{kq}{\sqrt{x^2 + (y+d)^2}}$



```python
#parametsr
k=1.
q=1.
d=1.

dx = .06
dy = .06
x,y = meshgrid(arange(-2.,2.,dx), arange(-2.,2.,dy))

Vxy = (k*q) / ((x**2 + (y-d)**2))**.5 - ((k*q) / (x**2 + (y + d)**2))**.5

from mpl_toolkits.mplot3d import Axes3D
fig = figure(1)
ax = subplot(1,1,1, projection = '3d')
p = ax.plot_surface(x, y, Vxy, rstride=1, cstride=1, linewidth=0, cmap=cm.BuPu)
#p = ax.plot_surface(x, y, Vxy, rstride=1, cstride=1, linewidth=0, cmap=cm.rainbow, vmin=-30, vmax=30)

cb = fig.colorbar(p, shrink=.5)
plt.show()
```

```python
levels = np.array([8,3,2,1,.1,0,-.1,-2,-5,-7,-8])
plt.contour(x,y, Vxy, levels, cmap=cm.coolwarm, linewidths=3)
plt.show()
```

\begin{equation} G(x,y) = e^{-x^2-y^2} \end{equation}

```python
dx = 12./51.
dy = 12./51.
x,y = meshgrid(arange(-6.,6.,dx), arange(-6., 6., dy))

G = e**(-x**2-y**2)

from mpl_toolkits.mplot3d import Axes3D
fig = figure(2)
ax = subplot(1,1,1, projection = '3d')
p = ax.plot_surface(x, y, G, rstride=1, cstride=1, linewidth=0, cmap=cm.BuPu)
#p = ax.plot_surface(x, y, Vxy, rstride=1, cstride=1, linewidth=0, cmap=cm.rainbow, vmin=-30, vmax=30)

cb = fig.colorbar(p, shrink=.5)
#offset = -6., cmap = cm.rainbow
plt.show()
```
