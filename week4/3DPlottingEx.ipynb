Electric Diapole Potential
====

$ V(x,y) = \displaystyle\frac{kq}{\sqrt{x^2 + (y-d)^2}} - \frac{kq}{\sqrt{x^2 + (y+d)^2}}$



```python
#params
k=1.
q=1.
d=1.

dx = .06
dy = .06
x,y = meshgrid(arange(-2.,2.,dx), arange(-2.,2.,dy))

Vxy = (k*q) / ((x**2 + (y-d)**2))**.5 - ((k*q) / (x**2 + (y + d)**2))**.5

from mpl_toolkits.mplot3d import Axes3D
fig = figure(1)
ax = subplot(1,1,1, projection = '3d')
p = ax.plot_surface(x, y, Vxy, rstride=1, cstride=1, linewidth=0, cmap=cm.coolwarm)

cb = fig.colorbar(p, shrink=.5)
plt.show()
```

Contour Plot

```python
levels = np.array([8, 4, 2, 1, 0.5, 0.25, 0, -0.25, -0.5, -1, -2, -4, -8])
plt.contour(x,y, Vxy, levels, cmap=cm.coolwarm, linewidths=3)
plt.show()
```

Ex 2. Quadrupole Potential
-----

Determine the electric potential for a quadrupole consisting of q= -1.0 nC at both (x =±2m,y=0m),andq=+1.0nCatboth(x=0m,y=±2m). Makeasurfaceplotand acontourplotontherange-4<(x,y)<4using101binsinxand101binsiny. Inthis case, do not let k = q = d = 1. Use dimensionful quantities and compute the value of the potential in volts.

```python
from mpl_toolkits.mplot3d import Axes3D
#setup the grid
dx = 8./101.
dy = 8./101.
psize = 4.
x,y = meshgrid(arange(-psize, psize, dx), arange(-psize, psize, dy))

#d is radial separation of the dipole
d = 2
q = 1E-9
#units of k: Nmm/CC
k = 8.987551787368176

psize = 4.
x,y = meshgrid(arange(-psize, psize, dx), arange(-psize, psize, dy))

#add two dipoles in one giant potential
Vxy = ((k*q) / (((x-0)**2 + (y-d)**2))**.5) + ((k*q) / (((x-0)**2 + (y+d)**2))**.5) - ((k*q) / (((x-d)**2 + (y-0)**2))**.5) - ((k*q) / (((x+d)**2 + (y-0)**2))**.5)

fig = figure()

ax = subplot(1,1,1, projection = '3d')
p = ax.plot_surface(x, y, Vxy, rstride=1, cstride=1, linewidth=0, cmap=cm.coolwarm)

cb = fig.colorbar(p, shrink=.5)
plt.show()
```

Ex 3. Potentials with Functions
------

Now we want to refactor the code we wrote in the example and first exercise to make it more useful. In this exercise, you will encapsulate your electric potential calculations within functions.

```python
from mpl_toolkits.mplot3d import Axes3D
#calculates the potential at a point or over an array due to a point particle
#in: float(s) x and y, the values to calculate the potential on
#   float q, the signed magnitude of the charge
#   float posx and posy, the location of the point charge
#out: float(s) of potential values

def pointPotential(x, y, q, posx, posy):
    return ((k*q) / (((x-posx)**2 + (y-posy)**2))**.5)


#calculates the potential at a point or over an array due to a dipole
#evenly spaces two point potentials about the origin
#in: float(s) x and y, the values to calculate the potential on
#   float q, the signed magnitude of the charge
#   float d, the radial distance separating the dipole
#   float rot, the angle of rotation in degrees to rotate the dipole about the orign
#       0 degrees is positive pole on the +x axis, CCW rotation
#out: float(s) of potential values

def dipolePotential(x, y, q, d, rot):
    #calibrate rotation so that 0 is positive dipole along x and the turn is ccw
    rot = -rot + 90
    #convert to radians
    rot = rot * 2* pi/360
    #rotate the positions of the point particles
    xrot = d*sin(rot)
    yrot = d*cos(rot)
    return pointPotential(x, y, q, xrot, yrot) - pointPotential(x, y, q, -xrot, -yrot)
```

#### Dipole Plot using Functions

```python
Vxy = dipolePotential(x, y, q, 1, 90)

fig = figure()
ax = subplot(1,1,1, projection = '3d')
p = ax.plot_surface(x, y, Vxy, rstride=1, cstride=1, linewidth=0, cmap=cm.coolwarm)
cb = fig.colorbar(p, shrink=.5)

plt.show()
```

#### Dipole Contour Plot using Functions

```python
levels = np.array([8, 4, 2, 1, 0.5, 0.25, 0, -0.25, -0.5, -1, -2, -4, -8])
levels = levels*1E-9
plt.contour(x,y, Vxy, levels, cmap=cm.coolwarm, linewidths=3)
plt.show()
```

#### Quadrupole Plot using Functions

```python
Vxy = pointPotential(x, y, -q, 2, 0) + pointPotential(x, y, -q, -2, 0) + pointPotential(x, y, q, 0, 2) + pointPotential(x, y, q, 0, -2)

fig = figure()
ax = subplot(1,1,1, projection = '3d')

p = ax.plot_surface(x, y, Vxy, rstride=1, cstride=1, linewidth=0, cmap=cm.coolwarm)
cb = fig.colorbar(p, shrink=.5)

plt.show()
```

#### Quadropole Contour Plot using Functions

```python
levels = np.array([8, 4, 2, 1, 0.5, 0.25, 0, -0.25, -0.5, -1, -2, -4, -8])
levels = levels*1E-9
plt.contour(x,y, Vxy, levels, cmap=cm.coolwarm, linewidths=3)
plt.show()
```
