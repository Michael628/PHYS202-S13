#### Exercise 2. Exponential function - numeric

```python
N = 10000
z = numpy.random.random_sample(N)
t = arange(.01, 5, .01)



cumul  = cumsum(f(t)) / (sum(f(t)))

ss = cumul.searchsorted(z)

hist(t[ss], 50, normed=True, label="Sampled Distribution")
plot(t, f(t), label="True Distribution", lw=3, color='red')
legend(loc='best')
show()
```

#### (c) Reusable function

```python
def distribute1D(t, prob, N):
    """takes any numerical distribution prob, on the interval defined by the array x,
    and returns an array of N sampled values that are statiscally the same as the input data."""
    
    N = 10000
    z = numpy.random.random_sample(N)
    
    cumul  = cumsum(prob) / (sum(prob))
    
    ss = cumul.searchsorted(z)
      
    return t[ss]


hist(distribute1D(t, f(t), N), 50, normed=True, label="Sampled Distribution")
#plot(t, f(t), label="True Distribution", lw=3, color='red')
legend(loc='best')
show()
```
