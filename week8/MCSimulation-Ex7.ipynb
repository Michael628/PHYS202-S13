#### 2. Weibull Distribution

$ F(x) = \displaystyle \int_0^x \alpha\beta^{-\alpha}t^{\alpha-1}e^{-(t/ \beta)^\alpha} dt$

$F(x) = 1 - e^{-(x/\beta)^\alpha}$

$ F^{-1}(x) = -\beta (\ln (1-x))^{1/\alpha}$

```python
def distribute1D(x, probfunc, N):
    """takes any numerical distribution prob, on the interval defined by the array x,
    and returns an array of N sampled values that are statiscally the same as the input data."""
    
    z = numpy.random.random_sample(N)
    cumul  = cumsum(probfunc) / (sum(probfunc))
    ss = cumul.searchsorted(z)
      
    return x[ss]
```

```python
alpha = 1.5
beta = 6

f = lambda x: alpha*beta**(-alpha) * x**(alpha-1) * exp(-(x/beta)**alpha)
F = lambda x: 1-exp(-(x/beta)**alpha)
Finv = lambda x: -beta*(log(1-x))**(1/alpha)

x = np.arange(0,25,.01)

N = 10000

n,b,p = hist(distribute1D(x, f(x), N), 100, normed=True, label="Sampled Distribution")
plot(x, F(x)*max(n), label="CDF", lw=3, color='green', linestyle="-", linewidth=2)
plot(x, f(x), label="True Distribution", lw=3, color='red', linestyle="--", linewidth=2)

legend(loc='upper right')
#axis("tight")
show()
```

#### 3. Beta(4,3) Distribution

$f(x) = 60 x^3 (1-x)^2, 0\leq x \leq 1$

$F(x) = 10x^4 (x^2 -2.4x +1.5) $

```python
f = lambda x: 60*x**3*(1-x)**2

F = lambda x: 10*x**4*(x**2 - 2.4*x +1.5)

x = np.arange(0,1.01,.01)
xmin =0
xmax = 1

#find maximum for bb
subx = x[logical_and((x > xmin), (x < xmax))]
ymax = max(f(subx))*1.05
ymin = 0.

N = 22000
samples_x = xmin + (xmax-xmin)*random_sample(N)
samples_y = ymin + (ymax - ymin) * random_sample(N)

myplot = plot(x,f(x), lw=3)
xlim(0.,1.)

newmask = (samples_y < f(samples_x))
sum(newmask)

#plot(samples_x[1:400], samples_y[1:400], 'g.')
plot(samples_x[newmask[:500]], samples_y[newmask[:500]], 'r.')
show()
print samples_x[newmask].size
```

```python
n,b,p = hist(samples_x[newmask], 100, normed=True, label="Sampled Distribution")
plot(x, F(x)*max(n), label="CDF", lw=3, color='green', linestyle="-", linewidth=2)
plot(x, f(x), label="True Distribution", lw=3, color='red', linestyle="--", linewidth=2)

xlim(0.,1)
legend(loc='lower right')
#axis("tight")
show()
```
