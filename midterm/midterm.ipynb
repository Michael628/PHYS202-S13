Anagram Strings

```python
def is_anagram(s1, s2):
    """
    checks if a string is an anagram
    in: string, string
    out: boolean
    """
    flag = True
    #count the characters in each string, an anagram will require that each character has the same
    #count in both strings
    for char in s1:
        if s1.count(char) != s2.count(char):
            flag = False
    return flag
    
print is_anagram("ab", "ba"), is_anagram("aa", "ba"), is_anagram("abddddd", "bad"), is_anagram("feeeddd", "dddeeef")
```
