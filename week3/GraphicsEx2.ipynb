2: Wave Function Graph

Graphs of wavefunction solutions and potential well

```python
#function parameters
A = 1
a = 1
k = 1
x = np.arange(-pi, pi+.01, .01)
#well function
u = -(1/cosh(a*x))**2

i = 1j
z = a*x

#psi squared function
#in: range, float
#out: function over range
def psi2(x, k):
    q = k/a
    return (q**2+tanh(z)**2)/(q**2 + 1)

#plot setup
fig = figure(1)
ax = subplot(111)     

plt.plot(x,u, linewidth = 1.5, label = "$U(x)$")

#generate the entire series of plots with different k parameters
#exclude k = 1.5
linestyles = ['-','--','-.', ' ', ':']
for m in range(0,5):
    n = float(m)/2
    if n != 1.5:
        plt.plot(x,psi2(x, n), linestyle = linestyles[m], color = 'red', linewidth = 1.5,  label = "$\psi^2,\ k=$"+str(n))
              
   
plt.legend(loc='best')
plt.axis([-pi-.3,pi+.3,-1.3,1.3])

plt.xlabel("$x(rad)$")
plt.ylabel("$U(x),\ \psi^2(x)$")

#hardwired pi labels on x axis
plt.xticks(np.arange(-pi, pi+.01, pi/2),[ "$-\pi$", "$-\pi/2$", "0", "$\pi/2$", "$\pi$"])
plt.yticks(np.arange(-1.0,1.01,.5) , np.arange(-1.0,1.01,.5))

plt.grid(True)
#unnecessary on chrome
#plt.figsize(9,7)

plt.show()
```

3: Sunspot Historical Data

```python
import re

#line by line parser
datalist = []
datain = open('sunspots.txt', 'r')
for line in datain:
    datalist.append(line)
datain.close()

#arrays to hold the sunspot data
ssall = np.zeros(len(datalist))
ssdates = np.zeros(len(datalist))
ssavg = np.zeros(len(datalist))

#hardwired startdate
startdate = 1749.0

#regexp based parser to pick out the data from the file input
for i, d in enumerate(datalist):
    #\t tab \d any digit + similar to * but requires one hit, works on preceding
    #r at the front forces the string to raw to help deal with chained \ escapes
    #total regexp query: match [tab]123.45
    query = r'\t\d+\.\d+'
    hit = re.search(query, str(d))
    #get the string the regexp search returns
    ssall[i] = float(hit.group())
    #end dates increase by 1/12 of a year at a time
    ssdates[i]  = startdate+(i+1)/12.
    
fig = figure(2)
ax = subplot(111)

plt.plot(ssdates, ssall, color = 'red', linewidth = .5, label = "Monthly Sunspot Count")
plt.legend(loc = 'best')

plt.xlabel("Year")
plt.ylabel("Sunspot Count")
plt.title("Historical Sunspot Activity")

#dynamic enddate, count hardwired
plt.axis([startdate,startdate+(len(ssdates)/12), 0, 300])
ax.minorticks_on()

plt.show()
```

3: b. Sunspots with running average.

Only the first 1000 datapoints analyzed.

```python
import re

#line by line parser
datalist = []
datain = open('sunspots.txt', 'r')
for i, line in enumerate (datain):
    datalist.append(line)
    #only process the first 1000 data points
    if i >=999:
        break
datain.close()

#arrays to hold the sunspot data
ssall = np.zeros(len(datalist))
ssdates = np.zeros(len(datalist))
ssavg = np.zeros(len(datalist))
sstot = 0

#regexp based parser to pick out the data from the file input
for i, d in enumerate(datalist):
    #\t tab \d any digit + similar to * but requires one hit, works on preceding
    #r at the front forces the string to raw to help deal with chained \ escapes
    #total regexp query: match [tab]123.45
    query = r'\t\d+\.\d+'
    hit = re.search(query, str(d))
    #get the string the regexp search returns
    ssall[i] = float(hit.group())
    #end dates increase by 1/12 of a year at a time
    ssdates[i]  = startdate+(i+1)/12.
    
    #calculate the running average
    sstot += float(hit.group())
    ssavg[i] = sstot/(i+1)
    
fig = figure(3)
ax = subplot(111)

plt.plot(ssdates, ssall, color = 'red', linewidth = .5, label = "Monthly Sunspot Count")
plt.plot(ssdates, ssavg, color = 'darkblue', linestyle = '-', label = "Running Average")
plt.legend(loc = 'best')

plt.xlabel("Year")
plt.ylabel("Sunspot Count")
plt.title("Historical Sunspot Activity")

#dynamic enddate, count hardwired
plt.axis([startdate,startdate+(len(ssdates)/12), 0, 250])
ax.minorticks_on()

plt.show()
```
