```python
import re
datalist = []

datain = open('sunspots.txt', 'r')
for line in datain:
    datalist.append(line)
datain.close()

ssall = np.zeros(len(datalist))
ssdates = np.zeros(len(datalist))
ssavg = np.zeros(len(datalist))

startdate = 1749.0
enddate = startdate

sstot = 0

for i, d in enumerate(datalist):
    #\t tab \d any digit + similar to * but requires one hit, works on preceding
    #r at the front forces the string to raw to help deal with chained \ escapes
    query = r'\t\d+\.\d+'
    hit = re.search(query, str(d))
    ssall[i] = float(hit.group())
    ssdates[i]  = startdate+(i+1)/12.
    sstot += float(hit.group())
    ssavg[i] = sstot/(i+1)
    
   # if i>1000:
     #   break

figure(1)
subplot(111)
plot(ssdates, ssall, color = 'red', linewidth = .5, label = "Monthly Sunspot Count")
plot(ssdates, ssavg, color = 'darkblue', linestyle = '-', label = "Cumulative Average")
legend(loc = 'best')
xlabel("Year")
ylabel("Sunspot Count")
title("Historical Sunspot Activity")
axis([startdate,startdate+(len(ssdates)/12), 0, 300])

ax = plt.gca()
minorticks_on()
#formatter = mtick(unit='Hz', places=1)
#ax.xaxis.set_major_formatter(formatter)


plt.show()
```
