```python
#create a simple plot
t = arange(0., 1.+01, .01)
c = cos(2*2*pi*t)
s = sin(4*pi*t)

#example of using latex commands
plot(t,c, color = 'blue', linewidth = 2, linestyle = ':', label = "$cos(4{\pi}t)$")
plot(t,s, color = 'red', linewidth = .5 , label = "$sin(4{\pi}t)$")
xlabel('time (s)')
ylabel("voltage (mV)")
title("AC Voltage", fontsize = 20)
grid(True)
legend(loc = "lower left")
savefig('quickplot.pdf')
```
